```python
#Implementation of Stack
class stack:
  def __init__(self,size):
    self.top=-1
    self.size=size
    self.s=[None]*size
  def isempty(self):
    return self.top==-1
  def isfull(self):
    return self.top+1==self.size
  def push(self,ele):
    if self.isfull():
      print("stack overflow")
    else:
      self.top+=1
      self.s[self.top]=ele
  def pop(self):
    if self.isempty():
      print("stack underflow")
    else:
      t=self.s[self.top]
      self.top=-1
      return t
  def top(self):
    if self.isempty():
      print("stack underflow")
    else:
      return self.s[self.top]
  def len_s(self):
    return self.top+1
  def print_(self):
    for i in range(self.top+1):
      print(self.s[i],end=" ")
    print("\n")
x=stack(5)
x.push(20)
x.push(21)
x.push(22)
x.push(23)
x.print_()
x.pop()
x.push(2)
x.print_()
x.len_s()
x.print_()
```
